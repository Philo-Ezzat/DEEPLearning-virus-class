```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import (Conv2D, BatchNormalization, Activation, Add, 
                                     GlobalAveragePooling2D, Dense, Input, MaxPooling2D)
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, confusion_matrix
import cv2
import os

# Define a residual block for ResNet-like architecture
def residual_block(x, filters, kernel_size=(3, 3), stride=1):
    shortcut = x  # Save the input for the shortcut connection

    # First convolutional layer
    x = Conv2D(filters, kernel_size, strides=stride, padding="same")(x)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)

    # Second convolutional layer
    x = Conv2D(filters, kernel_size, strides=1, padding="same")(x)
    x = BatchNormalization()(x)

    # Adjust the shortcut connection if dimensions do not match
    if stride != 1 or shortcut.shape[-1] != filters:
        shortcut = Conv2D(filters, (1, 1), strides=stride, padding="same")(shortcut)
        shortcut = BatchNormalization()(shortcut)

    # Add the shortcut connection and apply ReLU activation
    x = Add()([x, shortcut])
    x = Activation("relu")(x)
    return x

# Build a ResNet-like model
def build_resnet(input_shape, num_classes):
    inputs = Input(shape=input_shape)

    # Initial convolutional layer
    x = Conv2D(64, (7, 7), strides=2, padding="same")(inputs)
    x = BatchNormalization()(x)
    x = Activation("relu")(x)
    x = MaxPooling2D((3, 3), strides=2, padding="same")(x)

    # Add residual blocks
    x = residual_block(x, 64)  # Block with 64 filters
    x = residual_block(x, 128, stride=2)  # Block with 128 filters and downsampling
    x = residual_block(x, 256, stride=2)  # Block with 256 filters and downsampling
    x = residual_block(x, 512, stride=2)  # Block with 512 filters and downsampling

    # Global average pooling and fully connected layers
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation="relu")(x)
    outputs = Dense(num_classes, activation="softmax")(x)

    # Create the model
    model = Model(inputs, outputs)
    return model

# Function to preprocess images
def preprocess_images(image_path, target_size=(224, 224), brightness_factor=1.2, gamma=1.0):

    # Load the image
    img = cv2.imread(image_path)
    
    if img is None:
        raise ValueError(f"Image at path {image_path} could not be loaded.")
    
    # Resize the image to target size
    img_resized = cv2.resize(img, target_size)
    
    # Convert the image to float32 for processing
    img_resized = img_resized.astype(np.float32)
    
    # Apply brightness adjustment
    img_brightness = cv2.convertScaleAbs(img_resized, alpha=brightness_factor, beta=0)
    
    # Apply gamma correction
    img_gamma = np.power(img_brightness / 255.0, gamma) * 255.0
    img_gamma = np.clip(img_gamma, 0, 255).astype(np.uint8)
    
    # Normalize pixel values to [0, 1]
    img_normalized = img_gamma / 255.0
    
    return img_normalized

# Example of preprocessing all images in a directory
def preprocess_dataset(dataset_directory, target_size=(224, 224), brightness_factor=1.2, gamma=1.0):

    processed_images = []
    filenames = os.listdir(dataset_directory)
    
    for filename in filenames:
        image_path = os.path.join(dataset_directory, filename)
        
        # Preprocess the image
        processed_img = preprocess_images(image_path, target_size, brightness_factor, gamma)
        
        processed_images.append(processed_img)
    
    return np.array(processed_images)

# Model configuration
input_shape = (224, 224, 3)  # Image input dimensions (height, width, channels)
num_classes = 20  # Number of classes for classification

# Instantiate the model
model = build_resnet(input_shape, num_classes)

# Print the model summary
model.summary()

# Compile the model
model.compile(optimizer="adam", loss="categorical_crossentropy", metrics=["accuracy"])

# Data preparation with brightness adjustment and augmentation
train_datagen = ImageDataGenerator(
    rescale=1.0/255,  # Normalize pixel values to range [0, 1]
    rotation_range=30,  # Random rotation
    width_shift_range=0.2,  # Horizontal shift
    height_shift_range=0.2,  # Vertical shift
    shear_range=0.2,  # Shear transformation
    zoom_range=0.2,  # Random zoom
    horizontal_flip=True,  # Random horizontal flip
    fill_mode="nearest",  # Fill missing pixels after transformations
    brightness_range=[0.5, 1.5]  # Random brightness adjustment
)

# Validation data generator without augmentation
validation_datagen = ImageDataGenerator(rescale=1.0/255)  # Normalize pixel values to range [0, 1]

# Load training data from directory
train_generator = train_datagen.flow_from_directory(
    'data/train',  # Path to training data
    target_size=(224, 224),  # Resize images to match model input size
    batch_size=32,  # Number of images per batch
    class_mode='categorical'  # Output labels in one-hot encoded format
)

# Load validation data from directory
validation_generator = validation_datagen.flow_from_directory(
    'data/validation',  # Path to validation data
    target_size=(224, 224),  # Resize images to match model input size
    batch_size=32,  # Number of images per batch
    class_mode='categorical'  # Output labels in one-hot encoded format
)

# Train the model
history = model.fit(
    train_generator,
    epochs=10,  # Number of epochs
    validation_data=validation_generator  # Validation dataset
)

# Save the trained model
model.save('virus_classification_resnet_scratch.h5')



# Evaluation Metrics
# Get true labels and predictions

val_labels = validation_generator.classes
val_predictions = model.predict(validation_generator)
val_pred_classes = np.argmax(val_predictions, axis=1)

# Confusion Matrix
cm = confusion_matrix(val_labels, val_pred_classes)
plt.figure(figsize=(10, 8))
plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.colorbar()
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.xticks(np.arange(len(validation_generator.class_indices)), list(validation_generator.class_indices.keys()), rotation=45)
plt.yticks(np.arange(len(validation_generator.class_indices)), list(validation_generator.class_indices.keys()))
plt.show()


# Calculate metrics manually
accuracy = np.sum(np.diag(cm)) / np.sum(cm)  # Calculate accuracy
precision = np.diag(cm) / np.sum(cm, axis=0)  # Calculate precision for each class
recall = np.diag(cm) / np.sum(cm, axis=1)  # Calculate recall for each class
f1_scores = 2 * (precision * recall) / (precision + recall)  # Calculate F1-score for each class

# Average metrics
precision_avg = np.nanmean(precision)  # Average precision
recall_avg = np.nanmean(recall)  # Average recall
f1_avg = np.nanmean(f1_scores)  # Average F1-score

# Print evaluation metrics
print(f"Final Accuracy: {accuracy:.2f}")
print(f"Precision (avg): {precision_avg:.2f}")
print(f"Recall (avg): {recall_avg:.2f}")
print(f"F1 Score (avg): {f1_avg:.2f}")

# ROC Curve and AUC
val_labels_onehot = tf.keras.utils.to_categorical(val_labels, num_classes=len(validation_generator.class_indices))  # One-hot encode true labels
fpr, tpr, roc_auc = {}, {}, {}

# Compute ROC curve and AUC for each class
for i in range(len(validation_generator.class_indices)):
    fpr[i], tpr[i], _ = roc_curve(val_labels_onehot[:, i], val_predictions[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Generate distinct colors for all classes
num_classes = len(validation_generator.class_indices)
colors = plt.cm.tab20(np.linspace(0, 1, num_classes))  # Distinct colors using 'tab20' colormap

# Plot ROC Curve
plt.figure(figsize=(12, 10))
for i in range(num_classes):
    plt.plot(
        fpr[i], tpr[i], color=colors[i], lw=2,
        label=f'Class {list(validation_generator.class_indices.keys())[i]} (AUC = {roc_auc[i]:.2f})'
    )
plt.plot([0, 1], [0, 1], 'k--', lw=2)
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for All Classes')
plt.legend(loc="lower right", bbox_to_anchor=(1.35, 0.5), fontsize='small')
plt.show()
```

```python
import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.layers import (Conv2D, BatchNormalization, Activation, Add, 
                                     GlobalAveragePooling2D, Dense, Input, MaxPooling2D)
from tensorflow.keras.applications import DenseNet121
from tensorflow.keras.models import Model
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.xception import preprocess_input
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, confusion_matrix

# -------------------- Preprocessing Function --------------------

def preprocess_images(image_path, target_size=(224, 224), brightness_factor=1.2, gamma=1.0):
    """
    Preprocess the images by resizing, brightness adjustment, and gamma correction.
    
    :param image_path: Path to the image to preprocess.
    :param target_size: Target size for resizing (default: 224x224).
    :param brightness_factor: Factor for adjusting brightness (default: 1.2).
    :param gamma: Gamma correction factor (default: 1.0).
    :return: Preprocessed image as a numpy array.
    """
    # Load the image
    img = cv2.imread(image_path)
    
    if img is None:
        raise ValueError(f"Image at path {image_path} could not be loaded.")
    
    # Resize the image to target size
    img_resized = cv2.resize(img, target_size)
    
    # Convert the image to float32 for processing
    img_resized = img_resized.astype(np.float32)
    
    # Apply brightness adjustment
    img_brightness = cv2.convertScaleAbs(img_resized, alpha=brightness_factor, beta=0)
    
    # Apply gamma correction
    img_gamma = np.power(img_brightness / 255.0, gamma) * 255.0
    img_gamma = np.clip(img_gamma, 0, 255).astype(np.uint8)
    
    # Normalize pixel values to [0, 1]
    img_normalized = img_gamma / 255.0
    
    return img_normalized


# -------------------- Model Setup --------------------

def build_model():
    """
    Build and return the DenseNet121 model with custom classification layers.
    
    :return: Keras model instance.
    """
    base_model = DenseNet121(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

    # Add custom layers for classification
    x = base_model.output
    x = GlobalAveragePooling2D()(x)
    x = Dense(1024, activation='relu')(x)
    predictions = Dense(20, activation='softmax')(x)

    # Create the final model
    model = Model(inputs=base_model.input, outputs=predictions)

    # Freeze the base model layers
    for layer in base_model.layers:
        layer.trainable = False

    # Compile the model
    model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

    return model


# -------------------- Data Preparation Function --------------------

def prepare_data(train_dir, validation_dir):
    """
    Prepares training and validation data generators.
    
    :param train_dir: Directory containing training data.
    :param validation_dir: Directory containing validation data.
    :return: Train and validation data generators.
    """
    train_datagen = ImageDataGenerator(
        preprocessing_function=preprocess_input,
        rotation_range=30,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest'
    )

    validation_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

    train_generator = train_datagen.flow_from_directory(
        train_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical'
    )

    validation_generator = validation_datagen.flow_from_directory(
        validation_dir,
        target_size=(224, 224),
        batch_size=32,
        class_mode='categorical'
    )

    return train_generator, validation_generator


# -------------------- Train Model Function --------------------

def train_model(model, train_generator, validation_generator, epochs=10):
    """
    Train the model with the provided data generators.
    
    :param model: Keras model instance.
    :param train_generator: Data generator for training data.
    :param validation_generator: Data generator for validation data.
    :param epochs: Number of epochs for training.
    :return: History object containing training information.
    """
    history = model.fit(
        train_generator,
        epochs=epochs,
        validation_data=validation_generator
    )
    
    return history


# -------------------- Fine-Tune Model Function --------------------

def fine_tune_model(model, train_generator, validation_generator, epochs=10, learning_rate=0.0001):

    for layer in model.layers:
        layer.trainable = True

    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), 
                  loss='categorical_crossentropy', 
                  metrics=['accuracy'])

    history_fine = model.fit(
        train_generator,
        epochs=epochs,
        validation_data=validation_generator
    )

    return history_fine


# -------------------- Evaluation and Metrics --------------------

def evaluate_model(model, validation_generator):
    """
    Evaluate the model and compute evaluation metrics such as confusion matrix, accuracy, precision, recall, F1 score, and ROC-AUC.
    
    :param model: Trained Keras model instance.
    :param validation_generator: Data generator for validation data.
    :return: None
    """
    val_labels = validation_generator.classes
    val_predictions = model.predict(validation_generator)
    val_pred_classes = np.argmax(val_predictions, axis=1)

    # Confusion Matrix
    cm = tf.math.confusion_matrix(labels=val_labels, predictions=val_pred_classes).numpy()
    plt.figure(figsize=(10, 8))
    plt.imshow(cm, interpolation='nearest', cmap=plt.cm.Blues)
    plt.title("Confusion Matrix")
    plt.colorbar()
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    plt.xticks(range(len(validation_generator.class_indices)), validation_generator.class_indices.keys(), rotation=45)
    plt.yticks(range(len(validation_generator.class_indices)), validation_generator.class_indices.keys())
    plt.show()

    # Metrics
    accuracy = np.sum(np.diag(cm)) / np.sum(cm)
    precision = np.diag(cm) / np.sum(cm, axis=0)
    recall = np.diag(cm) / np.sum(cm, axis=1)
    f1_scores = 2 * (precision * recall) / (precision + recall)

    precision_avg = np.nanmean(precision)
    recall_avg = np.nanmean(recall)
    f1_avg = np.nanmean(f1_scores)

    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision_avg:.2f}")
    print(f"Recall: {recall_avg:.2f}")
    print(f"F1 Score: {f1_avg:.2f}")

    # ROC Curve and AUC
    val_labels_onehot = tf.keras.utils.to_categorical(val_labels, num_classes=len(validation_generator.class_indices))
    fpr, tpr, roc_auc = {}, {}, {}

    for i in range(len(validation_generator.class_indices)):
        fpr[i], tpr[i], _ = roc_curve(val_labels_onehot[:, i], val_predictions[:, i])
        roc_auc[i] = auc(fpr[i], tpr[i])

    plt.figure(figsize=(10, 8))
    colors = plt.cm.get_cmap('tab20', len(validation_generator.class_indices))

    for i, color in zip(range(len(validation_generator.class_indices)), colors.colors):
        plt.plot(fpr[i], tpr[i], color=color, lw=2,
                 label=f'Class {list(validation_generator.class_indices.keys())[i]} (AUC = {roc_auc[i]:.2f})')

    plt.plot([0, 1], [0, 1], 'k--', lw=2)
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    plt.show()


# -------------------- Main Execution --------------------

# Build model
model = build_model()

# Prepare data
train_generator, validation_generator = prepare_data('data/train', 'data/validation')

# Train model
train_history = train_model(model, train_generator, validation_generator, epochs=10)

# Fine-tune model
fine_tune_history = fine_tune_model(model, train_generator, validation_generator, epochs=10)

# Save model
model.save('virus_classification_densenet.h5')

# Evaluate model
evaluate_model(model, validation_generator)
```
